```python
# Fill null value
df['Income'].fillna(df['Income'].mean(),inplace=True)
df.isnull().sum()
```

```python
# Boxplot to detect outlier
figure=plt.figure(figsize=(22,8))
sns.boxplot(data=df)
plt.show()
```

```python
# Outlier Treatment using IQR Method
for col in df.columns:
  if df[col].dtype != 'object' and col != 'Dt_Customer':
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    IQR=q3-q1
    upper_bound=q3+1.5*IQR
    lower_bound=q1-1.5*IQR
    df[col]=np.where(df[col]>upper_bound,upper_bound,df[col])
    df[col]=np.where(df[col]<lower_bound,lower_bound,df[col])
print("Outlier Treated")
```

```python
# Boxplot after outlier treated
figure=plt.figure(figsize=(22,8))
sns.boxplot(data=df)
plt.show()
```

```python
# scatter plot to show the relationship between 'Income' and total spending on products.
df['Total Spending'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Income', y='Total Spending')
plt.title('Total Spending vs Income')
plt.xlabel('Income')
plt.ylabel('Total Spending')
plt.show()
```

```python
# Perform one-hot encoding on categorical columns
df = pd.get_dummies(df, columns=['Education', 'Marital_Status'], drop_first=True)
display(df.head())
```

```python
# Feature Engineering

# Calculate customer age
df['Age'] = 2024 - df['Year_Birth']

# Calculate total children
df['Total Children'] = df['Kidhome'] + df['Teenhome']

# Calculate customer tenure (in days)
df['Customer_Tenure'] = (pd.to_datetime('today') - df['Dt_Customer']).dt.days

display(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Select numerical columns excluding the datetime, target variable, and engineered features
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

# Assuming 'Response' is a target variable and should not be scaled
if 'Response' in numerical_cols:
    numerical_cols.remove('Response')

# Remove the engineered features from the list of numerical columns to scale
engineered_features = ['Age', 'Total Children', 'Customer_Tenure']
numerical_cols = [col for col in numerical_cols if col not in engineered_features]


# Instantiate StandardScaler
scaler = StandardScaler()

# Apply StandardScaler to the selected numerical columns
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

display(df.head())
```

# **Model Building**

```python
# Separate features (x) and target variable (y)
x = df.drop('Response', axis=1)
y = df['Response']
```

```python
from sklearn.model_selection import train_test_split

# Split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Elbow method(No. of cluster find)
from sklearn.cluster import KMeans
wcss=[]
# Drop the 'Dt_Customer' column as KMeans cannot handle datetime objects
df_numeric = df.drop('Dt_Customer', axis=1)
for i in range(2,21):
  km=KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10) # Added random_state and n_init for reproducibility and to suppress warning
  km.fit(df_numeric)
  wcss.append(km.inertia_)
```

```python
# Create a line plot across Wcss and no. of cluster
plt.plot([i for i in range(2,21)],wcss, marker='o')
plt.xlabel('No. of cluster')
plt.ylabel('Wcss')
plt.xticks([i for i in range(2,21)])
plt.grid(axis="x")
plt.show()
```

```python
# Perform K-Means with the best number of clusters (based on the Elbow method, which appears to be 3 from the previous plot)
kmn = KMeans(n_clusters=3, init='k-means++', random_state=42, n_init=10)

# Drop the 'Dt_Customer' column as KMeans cannot handle datetime objects
df_numeric = df.drop('Dt_Customer', axis=1)

# Fit the model and predict the clusters
kmn.fit(df_numeric)
df["Predict"]=kmn.predict(df_numeric)
display(df.head())
```

```python
from sklearn.metrics import silhouette_score

# Calculate the silhouette score
score = silhouette_score(df_numeric, df['Predict'])

# Print the silhouette score
print(f'Silhouette Score: {score}')
```

```python
# Plot the silhouette score
plt.figure(figsize=(6, 4))
plt.bar(['Silhouette Score'], [score], color='skyblue')
plt.ylabel('Score')
plt.title('Silhouette Score')
plt.ylim(0, 1)  # Silhouette score is between -1 and 1
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score, calinski_harabasz_score

# Calculate Davies-Bouldin Index
davies_bouldin = davies_bouldin_score(df_numeric, df['Predict'])
print(f'Davies-Bouldin Index: {davies_bouldin}')

# Calculate Calinski-Harabasz Index
calinski_harabasz = calinski_harabasz_score(df_numeric, df['Predict'])
print(f'Calinski-Harabasz Index: {calinski_harabasz}')
```

```python
# Visualize the clusters using a scatter plot of Income vs. Total Spending
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Income', y='Total Spending', hue='Predict', palette='viridis', s=50)
plt.title('K-Means Clustering Results (Income vs Total Spending)')
plt.xlabel('Income')
plt.ylabel('Total Spending')
plt.show()
```

```python
# Analyze the characteristics of each cluster
cluster_characteristics = df.groupby('Predict').agg({
    'Income': 'mean',
    'Age': 'mean',
    'Total Children': 'mean',
    'Total Spending': 'mean',
    'MntWines': 'mean',
    'MntFruits': 'mean',
    'MntMeatProducts': 'mean',
    'MntFishProducts': 'mean',
    'MntSweetProducts': 'mean',
    'MntGoldProds': 'mean',
    'NumDealsPurchases': 'mean',
    'NumWebPurchases': 'mean',
    'NumCatalogPurchases': 'mean',
    'NumStorePurchases': 'mean',
    'NumWebVisitsMonth': 'mean',
    'Customer_Tenure': 'mean',
    'Response': 'mean'
}).reset_index()

print("\n Kmeans Clustering - Cluster Characteristics:")
display(cluster_characteristics)
```
